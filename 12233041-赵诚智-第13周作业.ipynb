Make a Panel app to visualize the codon usage data

Optional : deploy as a local-server app and a web-server app



```python
import pandas as pd
import hvplot.pandas
import panel as pn
df = pd.read_csv('codon_usage.csv')
select = pn.widgets.Select(name='Select',options=df['triple'].unique().tolist())
@pn.depends(select)
def make_plot(select):
    df_plot = df[df['triple']==select]
    plot=df_plot.hvplot(kind='bar',x='codon',y='relative_frequency',xlabel='codon',ylabel='relative_frequency',width=400,title=select)
    return plot
pn.Column(pn.WidgetBox(select),make_plot)
```
